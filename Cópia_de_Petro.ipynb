# Análise de Produção de Petróleo no ano de 2021
 * Dados Públicos

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

```

```python
dados = pd.read_csv('/content/dadospetro.csv', sep = ';')
dados.head()
```

```python
dados.shape
```

```python
dados.describe
```

```python
dados['PRODUÇÃO DE PETRÓLEO'] = dados['PRODUÇÃO DE PETRÓLEO'].apply(lambda x: str(x).replace(",","."))
```

```python
dados['PRODUÇÃO DE PETRÓLEO']
```

```python
dados['PRODUÇÃO DE PETRÓLEO'].astype('float64')
```

```python
dados = dados[['PAÍS','ANO','REGIÃO GEOGRÁFICA','PRODUÇÃO DE PETRÓLEO']]
dados.head(10)
```

```python
dados
```

```python
dados_brasil = dados_brasil = dados.loc[dados['PAÍS']=='Brasil']

```

## Dados produção de Petróleo do Brasil 2021

```python
dados_brasil
```

# Análise de 2012 a 2021 do total de produção de petróleo no Brasil

```python
fig = px.line(dados_brasil, x = 'ANO', y = 'PRODUÇÃO DE PETRÓLEO', color = 'PAÍS')
fig.show()
```

# Dados produção de Petróleo da América do Norte
